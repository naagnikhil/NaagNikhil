#### Treating Outliers

```python
# Treat outliers in numerical columns using IQR method
numerical_cols = ['Shipment_Delay', 'Planned_TimeofTravel', 'Distance','Actual_Shipment_Time','Planned_Shipment_Time','Planned_Delivery_Time']
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```
